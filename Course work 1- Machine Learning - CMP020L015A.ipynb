```python
# X-axis and Y-axis

from sklearn.model_selection import train_test_split 
x = df.drop(['y'],axis=1) 
y = df['y']

type(x)

x
```

```python
# Spliting the data into training and testing set

X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=.20, random_state=0)

X_train
```

```python
# Using Linear Regression

from sklearn.linear_model import LinearRegression
linear=LinearRegression()
linear.fit(X_train,Y_train)
```

```python
y_pred=linear.predict(X_test)
print(y_pred)
```

```python
linear.predict([[1.68,82.8,24,6.554,6.538,6.438,6.390,6.318,29.44,39.83,59.10,54.11,40.72]])
```

```python
# Using Lasso Model

from sklearn.linear_model import Lasso
lasso_model = Lasso(alpha=1.0)
lasso=lasso_model.fit(X_train , Y_train)

# Prediction

lasso_predict = lasso.predict(X_test)
```

```python
import numpy as np
def PREDICT(Y_actual,Y_Predicted):
    mape = np.mean(np.abs((Y_actual - Y_Predicted)/Y_actual))*100
    return mape
```

```python
# Evaluation of the performance

Performance = PREDICT(Y_test,lasso_predict)
print("Value: ",Performance)
```

```python
# Evaluation of the performance - Accuracy

Accuracy = 100 - Performance
print('Accuracy of Lasso Regression: {:0.1f}%.'.format(Accuracy))
```

```python
# Visualisation of predictions results

plt.scatter(Y_test, lasso_predict)

# Title label
plt.title("Prediction Graph")
 
# X and Y labels
plt.xlabel('Y_test')
plt.ylabel('lasso_predict')

plt.show()
```
